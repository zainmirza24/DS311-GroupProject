```python
df.groupby('neutral_location')['country'].count() #T/F games played in neutral location 
```

```python
#top home team mean goals scored
goals_h = df.groupby('home_team')['home_team_score'].mean()
df_goalsH = goals_h.to_frame()
df_goalsH.sort_values(by = 'home_team_score', ascending = False).head(10)
```

```python
homeO = df.groupby('home_team')['home_team_mean_offense_score'].mean()
df_homeO = homeO.to_frame()
df_home_o = df_homeO.sort_values(by = 'home_team_mean_offense_score', ascending = False).head(10)
#TOP HOME TEAM OFFENSE 
df_home_o.round(2)
```

```python
goals_a = df.groupby('away_team')['away_team_score'].mean()
df_goalsA = goals_a.to_frame()
df_goalsA.sort_values(by = 'away_team_score', ascending = False).head(10)
#top away team mean goals scored
```

```python
awayO = df.groupby('away_team')['away_team_mean_offense_score'].mean()
df_awayO = awayO.to_frame()
df_away_o = df_awayO.sort_values(by = 'away_team_mean_offense_score', ascending = False).head(10)
#TOP AWAY TEAM OFFENSE 
df_away_o.round(2)
```

```python
#TOP OVerall Offensive  
offense = (df_home_o['home_team_mean_offense_score'] + df_away_o['away_team_mean_offense_score'])/2
df_o = offense.to_frame()
df_o.rename(columns = {0: "overall offense average"}).sort_values(by = "overall offense average", ascending = False).round(2)
```

# Away teams with their corresponding average away defense score

```python
awayd = df.groupby('away_team')['away_team_mean_defense_score'].mean() 
df_away = awayd.to_frame() #Series to Dataframe
```

# Top 10


```python
df_awayd = df_away.sort_values(by = 'away_team_mean_defense_score', ascending = False).head(10)
#Top 10 away defensive rating teams
```

```python
homed = df.groupby('home_team')['home_team_mean_defense_score'].mean() 
df_home = homed.to_frame() #Series to Dataframe
```

```python
df_homed = df_home.sort_values(by = 'home_team_mean_defense_score',ascending = False).head(10)
#Top 10 home defensive rating teams
```

```python
#Top overall defensive 
defense = (df_homed['home_team_mean_defense_score'] + df_awayd['away_team_mean_defense_score'])/2
df_d = defense.to_frame()
df_d.rename(columns = {0: "overall defense average"}).sort_values(by = "overall defense average", ascending = False).round(2)
```
