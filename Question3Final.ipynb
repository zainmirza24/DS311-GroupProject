```python
df_no_neutral = dataset.drop(dataset[dataset['neutral_location'] == True].index, inplace=False)
```

```python
df_no_neutral
```

```python
df_no_neutral_or_shoot_out = df_no_neutral[df_no_neutral['shoot_out'] == 'No']
```

```python
df_no_neutral_or_shoot_out
```

```python
df_no_neutral_or_shoot_out['home_team_result'].value_counts()
```

When cleaning up the data by filtering out the games played on neutral locations and ending in a shoot out we can see the matches played drop from 23,921 to 17,801. The home team wins 50.99% (9,077), loses 25.91% (4,614), and draws 23.11% (4,110) of the matches.

```python
df_no_neutral_or_shoot_out['home_team_result'].hist()
```

```python
home_team_wins_by_country = df_no_neutral_or_shoot_out['home_team_result'].groupby(df_no_neutral_or_shoot_out['home_team']).count()
```

```python
home_team_wins_by_country
```

```python
home_wins = df_no_neutral_or_shoot_out.groupby('home_team')['home_team_result'].count()
df_home = home_wins.to_frame()
```

```python
df_home.sort_values(by = 'home_team_result', ascending = False).head(30)
```

```python
away_wins = df_no_neutral_or_shoot_out.groupby('away_team')['home_team_result'].count()
df_away = away_wins.to_frame()
```

```python
df_away.sort_values(by = 'home_team_result', ascending = False).head(30)
```

```python
group_names = ["Win", "Lose", "Draw"]

#adding away team result column
def condition(x):
    if x == "Lose":
        return group_names[0]
    elif x == "Win":
        return group_names[1]
    else:
        return group_names[2]
df_no_neutral_or_shoot_out["away team result"] = df_no_neutral_or_shoot_out["home_team_result"].map(condition)
df_no_neutral_or_shoot_out
```

```python
away_wins_df = df_no_neutral_or_shoot_out.groupby('away_team')['away team result'].count()
df_away_wins = away_wins_df.to_frame()
```

```python
df_away_wins.sort_values(by = 'away team result', ascending = False).head(30)
```
