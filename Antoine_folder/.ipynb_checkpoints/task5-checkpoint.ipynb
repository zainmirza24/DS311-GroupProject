```python
# libraries 

import pandas as pd 
import numpy as np 
import datetime
import matplotlib.pyplot as plt
# Loading file 

sport_data_to_load = "fifa_world_cup.csv"

# Read Sport Data file and store into Pandas Data FRame

sport_data = pd.read_csv(sport_data_to_load)

sport_data
```

TASK 5 - Which team has the longest hot streak? 

```python
teams = sport_data['home_team'].unique().tolist()
teams
```

THIS IS THE CODE TO FIGURE OUT THE LONGEST HOT STREAK

```python
cmp = 0 
teams = sport_data['home_team'].unique().tolist()
lcmp = []
mcmp = []
date = []
datestreak = []
c = 0

for x in teams : 
    for i in range(len(sport_data)):
        if ((x == sport_data.iloc[i,1]) and (sport_data.iloc[i, 16] == 'Win')) or (x == sport_data.iloc[i,2]) and ((sport_data.iloc[i,16] == 'Lose')) :
            cmp = cmp+1
            c = 0
        if (x != sport_data.iloc[i,1]) or (x != sport_data.iloc[i,2]):
            c = c+1
        if ((x == sport_data.iloc[i,1]) and (sport_data.iloc[i,16] == 'Lose')) or ((x == sport_data.iloc[i,2]) and (sport_data.iloc[i,16] == 'Win')) or ((x == sport_data.iloc[i,1]) and (sport_data.iloc[i, 16] == 'Draw')) or ((x == sport_data.iloc[i,2]) and (sport_data.iloc[i,16] == 'Draw')):
            lcmp.append(cmp)
            date.append(sport_data.iloc[i-c,0])
            cmp = 0 
            c = 0
      
    mcmp.append(max(lcmp))
    datestreak.append(date[lcmp.index(max(lcmp))])
    lcmp=[]
    date=[]

mc = mcmp
dats = datestreak
```

```python
print(max(mcmp))
print(teams[mcmp.index(max(mcmp))])
print(datestreak[mcmp.index(max(mcmp))])
```

THE FOLLOWING DF SHOWS THE TOP 10 WINNING STREAK OF THE DATASET 

```python
# Creating a df of the longest hot streak for each team
df = pd.DataFrame(list(zip(teams, mc, dats)), columns = ['Teams','Longest hot streak', 'Last win of the streak'])
df_ = df.sort_values(by = ['Longest hot streak'], ascending = False).head(10)
df_
```

PLOT 1 : TOP 10 LONGEST HOT STREAK 

```python
# Plotting the barchart of the evolution of hot streak over the time for Spain 

fig = plt.figure(figsize = (17, 7))
plt.bar(df_['Teams'], df_['Longest hot streak'])

# print values above bars
for index, value in enumerate(df_['Longest hot streak']):
    plt.text(index, value,
             str(value), fontsize = 15)
# Changing the fontsize of the axes

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)


# labels and title 
plt.xlabel('Teams',fontsize = 20)
plt.ylabel("Streak's number", fontsize = 20)
plt.title("Top 10 longest hot streak", fontsize = 25)
```

Task 5 result : 
We found that Spain is the team with the longest winning streak : 25 matchs. The last win of this streak happened on the 20th of June 2009 against South Africa.
