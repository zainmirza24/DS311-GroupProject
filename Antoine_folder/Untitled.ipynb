```python
# libraries 

import pandas as pd 
import numpy as np 
import datetime
# Loading file 

sport_data_to_load = "fifa_world_cup.csv"

# Read Sport Data file and store into Pandas Data FRame

sport_data = pd.read_csv(sport_data_to_load)
sport_data
```

```python
sport_data['datetime'] = pd.to_datetime(sport_data['date'])
sport_data['year'] = sport_data['datetime'].dt.year

sport_data['year']
```

```python
sport_data.groupby('home_team').count().sort_values(by='date', ascending = False)
```

```python
sport_data['home_team_fifa_rank'].sort_values(ascending = True)
DT = sport_data.groupby('home_team_fifa_rank').count().reset_index()


DT.sort_values(by='home_team_fifa_rank')
```
