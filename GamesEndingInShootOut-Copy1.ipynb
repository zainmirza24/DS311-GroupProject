I am not that big into soccer but every 4 years I will watch the World Cup and I look forward to shoot outs because of the high pressure. I do watch the Champions league as well but it seems like that the a lot of matches end in a shoot out so I wanted to see if there is a higher chance for a game to end in a shoot out in the World Cup vs other matches.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("fifa_world_cup.csv")
```

```python
dataset.columns
```

```python
fifa_world_cup = dataset[dataset['tournament'] == 'FIFA World Cup']
```

```python
fifa_world_cup
```

```python
Games_without_draw = fifa_world_cup[fifa_world_cup['home_team_result'] != 'Draw']
```

```python
Games_without_draw
```

```python
Games_without_draw['shoot_out'].value_counts()
```

When looking at World Cup games ending in shoot outs they end in a shoot out around 6.25% of the time.

```python
Non_world_cup = dataset[dataset['tournament'] != 'FIFA World Cup']
```

```python
Non_world_cup
```

```python
Non_Fifa_without_draw = Non_world_cup[Non_world_cup['home_team_result'] != 'Draw']
```

```python
Non_Fifa_without_draw
```

```python
Non_Fifa_without_draw['shoot_out'].value_counts()
```

In non World Cup games there are 1.7% of them that end in a shoot out. By comparing the data we can see that there is a significant increase in matches that end in shoot outs at the World Cup compared to other matches.
